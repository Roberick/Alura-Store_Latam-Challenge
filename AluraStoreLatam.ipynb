### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



#### En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Seleccionamos la columna 'Precio' de la tienda 1 y usamos .sum() para sumar sus valores.
ingreso_tienda1 = tienda['Precio'].sum()

# Imprimimos el resultado para la tienda 1.
print(f"El Ingreso Total de la Tienda 1 es: {ingreso_tienda1}")
```

```python
# Hacemos lo mismo con el DataFrame de la tienda 2.
ingreso_tienda2 = tienda2['Precio'].sum()

# Imprimimos el resultado para la tienda 2.
print(f"El Ingreso Total de la Tienda 2 es: {ingreso_tienda2}")
```

```python
#Hacemolo mismo con el DataFame de la tienda 3.
ingreso_tienda3 = tienda3['Precio'].sum()

#Imprimos el resultado para la tienda 3.
print(f"El Ingreso Total de la Tienda 3 es: {ingreso_tienda3}")
```

```python
# Hacemos lo mismo con el DataFrame de la tienda 4.
ingreso_tienda4 = tienda4['Precio'].sum()

# Imprimimos el resultado para la tienda 4.
print(f"El Ingreso Total de la Tienda 4 es: {ingreso_tienda4}")
```

```python
print("--- RESUMEN DE INGRESOS TOTALES ---")
print(f"Tienda 1: {ingreso_tienda1}")
print(f"Tienda 2: {ingreso_tienda2}")
print(f"Tienda 3: {ingreso_tienda3}")
print(f"Tienda 4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
# Agrupamos los datos de la tienda 1 por 'Categoría del Producto' y contamos cuántas ventas hay en cada grupo.
productos_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("--- Tienda 1: Cantidad de Productos Vendidos por Categoría ---")
print(productos_tienda1)
```

```python
# Repetimos el proceso para la tienda 2.
productos_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("\n--- Tienda 2: Cantidad de Productos Vendidos por Categoría ---")
print(productos_tienda2)
```

```python
# Repetimos el proceso para la tienda 3.
productos_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("\n--- Tienda 3: Cantidad de Productos Vendidos por Categoría ---")
print(productos_tienda3)
```

```python
# Repetimos el proceso para la tienda 4.
productos_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("\n--- Tienda 4: Cantidad de Productos Vendidos por Categoría ---")
print(productos_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Seleccionamos la columna 'Calificacion_cliente' de la tienda 1 y usamos .mean() para calcular el promedio.
calificacion_tienda1 = tienda['Calificación'].mean()

# Imprimimos el resultado para la tienda 1.
print(f"La Calificación Promedio de la Tienda 1 es: {calificacion_tienda1:.2f}")
```

```python
# Repetimos el proceso para el DataFrame de la tienda 2.
calificacion_tienda2 = tienda2['Calificación'].mean()

# Imprimimos el resultado para la tienda 2.
print(f"La Calificación Promedio de la Tienda 2 es: {calificacion_tienda2:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Agrupar por producto y contar las ventas, ordenando de mayor a menor.
conteo_productos_t1 = tienda.groupby('Producto').size().sort_values(ascending=False)

# Producto más vendido (el primero de la lista)
mas_vendido_t1 = conteo_productos_t1.head(1)

# Producto menos vendido (el último de la lista)
menos_vendido_t1 = conteo_productos_t1.tail(1)

print("--- Tienda 1 ---")
print("⭐ MÁS VENDIDO:")
print(mas_vendido_t1)
print("\n❌ MENOS VENDIDO:")
print(menos_vendido_t1)
```

```python
# Repetimos el proceso para la tienda 2.
conteo_productos_t2 = tienda2.groupby('Producto').size().sort_values(ascending=False)

mas_vendido_t2 = conteo_productos_t2.head(1)
menos_vendido_t2 = conteo_productos_t2.tail(1)

print("\n--- Tienda 2 ---")
print("⭐ MÁS VENDIDO:")
print(mas_vendido_t2)
print("\n❌ MENOS VENDIDO:")
print(menos_vendido_t2)
```

```python
# Repetimos el proceso para la tienda 3.
conteo_productos_t3 = tienda3.groupby('Producto').size().sort_values(ascending=False)

mas_vendido_t3 = conteo_productos_t3.head(1)
menos_vendido_t3 = conteo_productos_t3.tail(1)

print("\n--- Tienda 3 ---")
print("⭐ MÁS VENDIDO:")
print(mas_vendido_t3)
print("\n❌ MENOS VENDIDO:")
print(menos_vendido_t3)
```

```python
# Repetimos el proceso para la tienda 4.
conteo_productos_t4 = tienda4.groupby('Producto').size().sort_values(ascending=False)

mas_vendido_t4 = conteo_productos_t4.head(1)
menos_vendido_t4 = conteo_productos_t4.tail(1)

print("\n--- Tienda 4 ---")
print("⭐ MÁS VENDIDO:")
print(mas_vendido_t4)
print("\n❌ MENOS VENDIDO:")
print(menos_vendido_t4)
```

#6. Generación de Gráficos

###Gráfico de Barras: Ingreso Total por Tienda

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Recopilación de los datos de ingresos (Usando los resultados del paso 1)
datos_ingresos = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
}
df_ingresos = pd.DataFrame(datos_ingresos)

# 2. Creación del gráfico
plt.figure(figsize=(8, 5))
plt.bar(df_ingresos['Tienda'], df_ingresos['Ingreso Total'], color=['skyblue', 'lightgreen', 'salmon', 'gold'])

# 3. Etiquetas y título
plt.title('Ingreso Total Generado por Cada Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total Estimado (Unidad Monetaria)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###Gráfico de Puntos: Calificación Promedio de Clientes

```python
# 1. Recopilación de los datos de calificaciones (Usando los resultados del paso 3)
datos_calificaciones = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
}
df_calificaciones = pd.DataFrame(datos_calificaciones)

# 2. Creación del gráfico de puntos (o línea)
plt.figure(figsize=(8, 5))
plt.plot(df_calificaciones['Tienda'], df_calificaciones['Calificación Promedio'],
         marker='o', linestyle='-', color='indigo', markersize=10, linewidth=2)

# 3. Etiquetas y título
plt.title('Calificación Promedio del Cliente por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(df_calificaciones['Calificación Promedio'].min() - 0.1,
         df_calificaciones['Calificación Promedio'].max() + 0.1) # Ajuste del eje Y
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

###Gráfico de Torta (Pie Chart): Distribución de Categorías (Tienda 1)

```python
# 1. Seleccionar la top 5 de categorías más vendidas de la Tienda 1
# Nota: La variable productos_tienda1 ya se generó en el paso 2
top_5_categorias_t1 = productos_tienda1.head(5)

# 2. Creación del gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_categorias_t1,
    labels=top_5_categorias_t1.index,
    autopct='%1.1f%%', # Formato para mostrar porcentajes
    startangle=90,
    colors=plt.cm.Set3.colors, # Uso de un mapa de colores
    wedgeprops={'edgecolor': 'black'}
)

# 3. Título
plt.title('Distribución de Ventas por Categoría (Top 5) en Tienda 1')
plt.axis('equal') # Asegura que el gráfico de torta sea circular
plt.show()
```

#7. ¡Extra! Análisis del desempeño geográfico!


```python
# Lista de todos los DataFrames de las tiendas
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

# Combinar todos los DataFrames en un solo DataFrame maestro
ventas_consolidadas = pd.concat(lista_tiendas, ignore_index=True)

print("Datos consolidados listos. Número total de ventas:", len(ventas_consolidadas))
```

```python
import matplotlib.pyplot as plt

# Definimos el tamaño de la figura (opcional, pero mejora la vista)
plt.figure(figsize=(10, 8))

# Creamos el gráfico de dispersión:
# x=longitud, y=latitud
# s=tamaño del punto (lo hacemos proporcional al Precio, multiplicado por un factor para que se vea)
# c=color del punto (lo basamos en la Calificacion_cliente)
scatter = plt.scatter(
    ventas_consolidadas['lon'],             # Eje X: Longitud
    ventas_consolidadas['lat'],             # Eje Y: Latitud
    s=ventas_consolidadas['Precio'] / 10,   # Tamaño: Proporcional al Precio (dividido para escalar)
    c=ventas_consolidadas['Calificación'], # Color: Basado en la Calificación
    cmap='viridis',                         # Mapa de color
    alpha=0.6,                              # Transparencia
    edgecolors='w',                         # Borde blanco
    linewidths=0.5
)

# Añadimos etiquetas y título
plt.title('Distribución Geográfica de Ventas (Tamaño=Precio, Color=Calificación)')
plt.xlabel('Longitud (lon)')
plt.ylabel('Latitud (lat)')

# Añadimos una barra de color para entender las calificaciones
cbar = plt.colorbar(scatter)
cbar.set_label('Calificación')

# Mostramos el gráfico
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
